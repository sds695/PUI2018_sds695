`````python
def moving_average(a, n=3) :
    '''
    This function returns the numpy array with moving average applied.
    Input: numpy array, The number of elements on which the average applies
    Output: Array with moving average applied
    '''
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
`````

`````python
plt.figure(figsize=(14,10))
plt.xlabel("Date entry per week")
plt.ylabel("Number of swipes(log10)")
for i in range(23):
    plt.plot(date[:192],moving_average(np.log10(np.nanmean(mta,axis=0)[i])))
`````

Figure 2: Timeseries of number of swipes for each ride type. Moving average was calculated to smoothen the curve. There appears to be a big drop in the light blue line right before 2011-03.

`````python
for i in range(5):
    N = len(mta_filtered[i])
    f = np.abs(np.fft.rfft(mta_filtered[i]))**2
    fig=plt.figure(figsize=(15,5))
    ax=fig.add_subplot(111)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), 'o', ms=20)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-')

    # changing the labels on the x ticks to see the period instead of frequency
    ax.set_xticklabels([ "%s"%(1/f) for f in ax.get_xticks()], fontsize=20) 
    ax.axvline(x=1/52, color='r')
    plt.xlabel("period (week)", fontsize=20)
    plt.ylabel("Power", fontsize=20)
    plt.title("Trash pickup power spectrum", fontsize=20)
`````

To observe the periodicity i have plotted the squared fourier transform of the first 5 subway timeseries and also plotted a red line at the period that corresponds to 52 weeks(1 year). We can notice that the first and fourth graphs have a peak close to the line.
